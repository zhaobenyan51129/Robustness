```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def image_to_tensor_old(file_dir):
    """读取单张grating, return numpy"""
    with open(file_dir) as f:
        x_1 = np.fromfile(f, 'i4', 1)
        x_2 = np.fromfile(f, 'i4', 3)
        nFrame=x_2[0]
        size=x_2[1]
        x_3 = np.fromfile(f, 'f4', 3)
        x_4 = np.fromfile(f, 'f4', 2)
        x_5 = np.fromfile(f, 'u4', 1)
        x_6 = np.fromfile(f, 'f4', nFrame*size*size*3)
        image_numpy = x_6.reshape(size, size,3)
    return image_numpy
```

```python
def image_to_tensor(file_dir):
    """读取单张grating, return numpy"""
    with open(file_dir) as f:
        x_1 = np.fromfile(f, 'i4', 1)
        x_2 = np.fromfile(f, 'i4', 3)
        nFrame=x_2[0]
        size=x_2[1]
        x_3 = np.fromfile(f, 'f4', 3)
        x_4 = np.fromfile(f, 'f4', 2)
        x_5 = np.fromfile(f, 'u4', 1)
        x_6 = np.fromfile(f, 'f4', nFrame*size*size*3)
        image_numpy = x_6.reshape(3,size, size)
    return image_numpy.transpose([1,2,0]) # [224, 224, 3]
```

```python

file='/home/zhaobenyan/dataset/grating/grating_32x32_frameRate1/static_color-grid_5.bin'
image=image_to_tensor_old(file)
plt.imshow(image)
```

```python
image=image_to_tensor(file)
plt.imshow(image)
```
